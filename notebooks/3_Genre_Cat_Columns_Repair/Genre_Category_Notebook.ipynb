## Methods

```python
# this method is a cusom value checking
# returns true if input is not blank, 'data', or float
# returns false otherwise

def check_str(string):
    
    return_var = True
    
    # check if float
    if type(string) is float:
        return_var = False
    
    # check if blank
    elif string.isspace() == True:
        return_var = False
    
    # check if string == 'data'
    elif string == 'Data':
        return_var = False
    
    return return_var
```

```python
def genre_category_repair(df):
    
    for i, row in df.iterrows():
        # df.at[i,'category']
        # df.at[i,'genre']
        # df.at[i,''] = New_Track_id_counter
        # str.isalnum()
        
        # 1. get string for Category
        category = df.at[i,'category']
        
        # 2. get string for Genre
        genre = df.at[i,'genre']
        
        # 3. Condition Checking
        # 3a. Category = Data/blank, Genre = good data
        if check_str(category) == False and check_str(genre) == True:
            df.at[i,'category'] = genre
            print("3a. Before: " + str(category) + ", " + str(genre) + "; After: " + str(genre) + ", " + str(genre))
        
        # 3b. Categoy = good data, Genre = blank/bad
        elif check_str(category) == True and check_str(genre) == False:
            df.at[i,'genre'] = category
            print("3b. Before: " + str(category) + ", " + str(genre) + "; After: " + str(category) + ", " + str(category))
    
    return df
    
```

## running method

```python
clean_data = genre_category_repair(data)

# output below shows what was cleaned
```
